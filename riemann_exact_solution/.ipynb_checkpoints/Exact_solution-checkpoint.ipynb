```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma = 1.4
def p_star_func(px,py):
    if px >= py:
        val = 1.0/gamma*((px/py) - 1.0)/np.sqrt(((gamma+1.0)*px/(2.0*gamma*py)) + ((gamma-1.0)/(2.0*gamma)))
    else:
        val = 2/(gamma-1.0)*((px/py)**((gamma-1.0)/(2.0*gamma)) -1.0)
    return val

def p_der(px,py):
    if px >= py:
        xe = np.sqrt(((gamma+1.0)*px/(2.0*gamma*py)) + ((gamma-1.0)/(2.0*gamma)))
        val = (1.0/(gamma*py*xe)) - (((px/py) - 1.0)*(gamma+1.0)/(4.0*gamma*gamma*py*xe*xe*xe))
    else:
        val = 1.0/gamma*((px/py)**(-(gamma+1.0)/(2.0*gamma)))
    return val
```

```python
n = 100
def shocktube_exact(rho1,p1, u1, rho4,p4, u4,time,xval=None):
    a1 = np.sqrt(gamma*p1/rho1)
    a4 = np.sqrt(gamma*p4/rho4)
    p_star_o = (0.5*(p1+p4) - 0.125*(u4-u1)*(rho4+rho1)*(a4+a1))
    p_star = 0.0
    while(abs(p_star_o - p_star) > 1e-4):
        p_star = p_star_o
        if p_star_o < p1 and p_star_o < p4:
            z = (gamma-1.0)/(2.0*gamma)
            p_star = (((a1+a4) + (z*gamma*(u1-u4)))/(a1*p1**-z + a4*p4**-z))**(1.0/z)
            break
        p_star_o = p_star_o - (a1*p_star_func(p_star_o,p1)+ a4*p_star_func(p_star_o,p4) - u1 + u4)/(a1*p_der(p_star_o,p1)+ a4*p_der(p_star_o,p4))
    print("p_star = ",p_star_o )
    if p1 <= p_star and p4 <= p_star :
        pos1 = u1 - a1*np.sqrt(((gamma+1.0)/(2.0*gamma)*((p_star/p1) - 1.0)) + 1.0)
        pos2 = pos1
        pos4 = u4 + a4*np.sqrt(((gamma+1.0)/(2.0*gamma)*((p_star/p4) - 1.0)) + 1.0)
        pos3 = u1 - a1/gamma*((p_star/p1) - 1.0)/np.sqrt(((gamma+1.0)*p_star/(2.0*gamma*p1)) + ((gamma-1.0)/(2.0*gamma)))
        pos5 = pos4
        rho2 = rho1*(gamma-1.0 + ((gamma+1.0)*p_star/p1))/(gamma+1.0 + ((gamma-1.0)*p_star/p1))
        rho3 = rho4*(gamma-1.0 + ((gamma+1.0)*p_star/p4))/(gamma+1.0 + ((gamma-1.0)*p_star/p4))
    elif p1 <= p_star and p4 >= p_star :
        pos3 = u1 - a1/gamma*((p_star/p1) - 1.0)/np.sqrt(((gamma+1.0)*p_star/(2.0*gamma*p1)) + ((gamma-1.0)/(2.0*gamma)))
        pos1 = u1 - a1*np.sqrt(((gamma+1.0)/(2.0*gamma)*((p_star/p1) - 1.0)) + 1.0)
        pos2 = pos1
        pos4 = u4 + a4*(((gamma+1.0)/(gamma-1.0)*(p_star/p4)**((gamma-1)/(2.0*gamma))) - (2.0/(gamma-1.0)))
        pos5 = u4 + a4 
        rho2 = rho1*(gamma-1.0 + ((gamma+1.0)*p_star/p1))/(gamma+1.0 + ((gamma-1.0)*p_star/p1))
        rho3 = rho4*(p_star/p4)**(1.0/gamma)
    elif p1 >= p_star and p4 <= p_star :
        pos1 = u1 - a1
        pos2 = u1 + a1*((2.0/(gamma-1.0)) - ((gamma+1.0)/(gamma-1.0)*(p_star/p1)**((gamma-1)/(2.0*gamma))))
        pos3 = u4 + a4/gamma*((p_star/p4) - 1.0)/np.sqrt((gamma+1.0)*p_star/(2.0*gamma*p4) + ((gamma-1.0)/(2.0*gamma)))
        pos4 = u4 + a4*np.sqrt((gamma+1.0)/(2.0*gamma)*(p_star/p4 - 1.0) + 1.0)
        pos5 = pos4
        rho2 = rho1*(p_star/p1)**(1.0/gamma)
        rho3 = rho4*(gamma-1.0 + ((gamma+1.0)*p_star/p4))/(gamma+1.0 + ((gamma-1.0)*p_star/p4))
    elif p1 >= p_star and p4 >= p_star :
        pos1 = u1-a1
        pos2 = u1 + a1*((2.0/(gamma-1.0)) - ((gamma+1.0)/(gamma-1.0)*(p_star/p1)**((gamma-1)/(2.0*gamma))))
        pos3 = u1 + 2.0*a1/(gamma-1.0)*(1.0 - ((p_star/p1)**((gamma-1.0)/2.0/gamma)))
        pos4 = u4 + a4*(((gamma+1.0)/(gamma-1.0)*(p_star/p4)**((gamma-1)/(2.0*gamma))) - (2.0/(gamma-1.0)))
        pos5 = u4+a4
        rho2 = rho1*(p_star/p1)**(1.0/gamma)
        rho3 = rho4*(p_star/p4)**(1.0/gamma)
    xlim = 0.5
    if xval == None:
        x = np.linspace(-xlim,xlim,n)
    else:
        x = [xval]
    u_exact = [0.0]*len(x)
    p_exact = [0.0]*len(x)
    rho_exact = [0.0]*len(x)
    for i in range(len(x)):
        if x[i]< pos1*time and x[i]>= -xlim:
            u_exact[i] = u1
            p_exact[i] = p1
            rho_exact[i] = rho1
        elif x[i]<= pos2*time and x[i] >= pos1*time:
            u_exact[i] = 2.0/(gamma+1)*((x[i]/time) + ((gamma-1)/2.0*u1) + a1)
            p_exact[i] = p1*((gamma-1.0)/(gamma+1.0)/a1*(u1 - (x[i]/time)) + ((2.0)/(gamma+1.0)))**(2.0*gamma/(gamma-1.0))
            rho_exact[i] = rho1*(p_exact[i]/p1)**(1.0/gamma)
        elif x[i]<= pos3*time and x[i]>= pos2*time:
            u_exact[i] = pos3
            p_exact[i] = p_star
            rho_exact[i] = rho2
        elif x[i]<= pos4*time and x[i]>= pos3*time:
            u_exact[i] = pos3
            p_exact[i] = p_star
            rho_exact[i] = rho3
        elif x[i]<= pos5*time and x[i]>= pos4*time:
            u_exact[i] = 2.0/(gamma+1)*((x[i]/time) + ((gamma-1)/2.0*u4) -a4)
            p_exact[i] = p4*((gamma-1.0)/(gamma+1.0)/a4*((x[i]/time) -u4) + (2.0)/(gamma+1.0))**(2.0*gamma/(gamma-1.0))
            rho_exact[i] = rho4*(p_exact[i]/p4)**(1.0/gamma)
        elif x[i]<= xlim and x[i]>= pos5*time :
            u_exact[i] = u4
            p_exact[i] = p4
            rho_exact[i] = rho4
    return xlim,u_exact,p_exact,rho_exact
```

```python
def plot_shocktube(p1, u1,rho1,p4, u4, rho4,time):
    xlim,u_exact,p_exact,rho_exact = shocktube_exact(rho1,p1, u1, rho4,p4, u4,time)
    T_exact = [np.sqrt(gamma*p_exact[i]/rho_exact[i]) for i in range(len(p_exact))]
    x = np.linspace(-xlim,xlim,n)
    plt.figure()
    plt.plot(x,u_exact,label = 'exact')
    plt.xlabel('x')
    plt.ylabel('u')
    plt.xlim(-0.5,0.5)
    plt.title("Plot of velocity vs. x for Normal Shock tube Problem")
    plt.grid()
    plt.show()
    plt.figure()
    plt.xlim(-0.5,0.5)
    plt.xlabel('x')
    plt.ylabel('p')
    plt.title("Plot of pressure vs. x for Normal Shock tube Problem")
    plt.grid()
    plt.plot(x,p_exact,label = 'exact')
    plt.show()
    plt.figure()
    plt.xlabel('x')
    plt.xlim(-0.5,0.5)
    plt.grid()
    plt.ylabel('rho')
    plt.title("Plot of density vs. x for Normal Shock tube Problem")
    plt.plot(x,rho_exact,label = 'exact')
    plt.show()
    plt.figure()
    plt.plot(x,T_exact,label = 'exact')
    plt.title("Plot of Temperature vs. x for Normal Shock tube Problem")
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('T')
    plt.xlim(-0.5,0.5)
    plt.show()
```

```python
Test_cases = [(1.000,0.000,1.000 ,0.100 ,0.000,0.125 ,0.25),
              (0.400,-2.000,1.000 ,0.400,2.000,1.000 ,0.15 ),
              (1000.000,0.000,1.000 ,0.010,0.000,1.000 ,0.012), 
              (0.010,0.000,1.000 , 100.000 ,0.000,1.000, 0.035), 
              (460.894,19.5975,5.99924 , 46.095 , -6.19633,5.99242,0.035), 
              (1.000,0.750,1.000,0.100,0.000,0.125,0.20)
             ]
for b in range(len(Test_cases)):
    y = Test_cases[b]
    plot_shocktube(y[0],y[1],y[2],y[3],y[4],y[5],y[6])
```
